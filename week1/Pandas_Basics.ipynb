#PANDAS DAY 2

```python
import pandas as pd
import numpy as np
```

```python
employee_data = {
    "name": ["Arima", "Rahim", "Karim", "Jannat", "Sadia", "Nabil"],
    "department": ["CSE", "EEE", "CSE", "BBA", "CSE", "EEE"],
    "city": ["Dhaka", "Dhaka", "Khulna", "Dhaka", "Chattogram", "Khulna"],
    "salary": [50000, 45000, 60000, 40000, 55000, 47000]
}

employees_df = pd.DataFrame(employee_data)
print(employees_df)
```

```python
print(employees_df["department"].value_counts())
print(employees_df["city"].value_counts())
```

```python
print(employees_df["department"].unique())
print(employees_df["department"].nunique())
print(employees_df["city"].unique())
```

Group By Basics

```python
print(employees_df.groupby("department")["salary"].mean())
print(employees_df.groupby("department")["salary"].max())
print(employees_df.groupby("city")["salary"].mean())
```

```python
print(
    employees_df.groupby("department")["salary"].agg(["mean", "max", "min", "count"])
)
```

```python
employees_df.groupby(["department", "city"])["salary"].mean()
```

```python
employees_df.groupby("department").agg(
    {
        "salary":["mean", "count"]
    }
)
```

Transform

```python
employees_df["DeptMean"] = employees_df.groupby("department")["salary"].transform("mean")
print(employees_df)
```

Sort Grouped Result

```python
department_salary = employees_df.groupby("department")["salary"].mean()
print(department_salary.sort_values(ascending=False))
```

```python
df = pd.DataFrame({
    "name": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank"],
    "department": ["HR", "IT", "Finance", "IT", "HR", "Finance"],
    "city": ["Dhaka", "Chattogram", "Dhaka", "Khulna", "Dhaka", "Rajshahi"],
    "salary": [50000, 70000, 65000, 72000, 52000, 68000]
})

print("Original DataFrame:")
print(df)
```

```python
print("\ndepartment.value_counts():")
print(df["department"].value_counts())

print("\ncity.value_counts():")
print(df["city"].value_counts())
```

```python
print("Unique departments:")
print(df["department"].unique())

print("\nNumber of unique cities:")
print(df["city"].nunique())
```

```python
grouped = df.groupby("department").agg(
    average_salary=("salary", "mean"),
    maximum_salary=("salary", "max"),
    employee_count=("name", "count")
)

print("Grouped by department:")
print(grouped)
```

```python
grouped = df.groupby("department").agg(
    average_salary=("salary", "mean"),
    maximum_salary=("salary", "max"),
    employee_count=("name", "count")
).round(2)

print(grouped)
```

```python
messy_df = pd.DataFrame({
    "name": ["Alice", None, "Charlie", "David", None, "Frank"],
    "age": [25, 30, None, 28, 35, None],
    "salary": [50000, None, 65000, None, 52000, 68000]
})

print("Messy DataFrame:")
print(messy_df)
```

```python
print("\nisnull():")
print(messy_df.isnull())

print("\nisnull().sum():")
print(messy_df.isnull().sum())

print("\nisnull().any():")
print(messy_df.isnull().any())
```

```python
filled_df = messy_df.copy()

filled_df["name"] = filled_df["name"].fillna("Unknown")
filled_df["age"] = filled_df["age"].fillna(filled_df["age"].mean())
filled_df["salary"] = filled_df["salary"].fillna(filled_df["salary"].median())

print("DataFrame after filling missing values:")
print(filled_df)
```

```python
filled_df["age"] = filled_df["age"].round().astype(int)
print(filled_df)
```

```python
updated_df = df.rename(columns={"salary": "monthly_salary"}).drop(columns=["city"])

print("Updated DataFrame:")
print(updated_df)
```

##My understanding
1. What value_counts() does

value_counts() counts how many times each unique value appears in a column.

Example:
If the department column has HR, IT, HR, Finance, then value_counts() will tell us:

HR → 2

IT → 1

Finance → 1

It is useful when we want to quickly see the frequency of categories.

2. What groupby() does

groupby() groups rows that have the same value in a selected column, then lets us perform calculations on each group.

Example:
If we group by department, then all HR employees go together, all IT employees go together, and so on.
After that, we can calculate:

average salary

max salary

employee count

So groupby() is useful for summarizing data by category.

3. Difference between fillna() and dropna()

fillna() is used to replace missing values with something else.

Example:

missing name → "Unknown"

missing age → average age

dropna() is used to remove rows or columns that contain missing values.

So:

fillna() keeps the data and fills the gaps

dropna() removes incomplete data

4. When unique() is useful

unique() is useful when we want to know all distinct values in a column.

Example:
If the city column has Dhaka, Dhaka, Khulna, Rajshahi, then unique() gives:
['Dhaka', 'Khulna', 'Rajshahi']

It helps us understand what different categories exist in the data.
