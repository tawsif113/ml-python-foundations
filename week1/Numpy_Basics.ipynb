Create Special Arrays

```python
zeros = np.zeros((2, 3))
ones = np.ones((3, 2))
full = np.full((2, 2), 7)
identity = np.eye(3)

print(zeros)
print(ones)
print(full)
print(identity)
```

Indexing & Slicing

```python
range_arr = np.arange(0, 10, 2)
line_arr = np.linspace(0, 10, 5)

print(range_arr)
print(line_arr)
```

Vectorization

```python
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([10, 20, 30, 40])

print(arr1 + arr2)
print(arr2 - arr1)
print(arr1 * 2)
print(arr1 / 2)
print(arr1 ** 2)
print(arr1 + 5)
```

Aggregation

```python
arr = np.array([5, 10, 15, 20])

print(np.sum(arr))
print(np.mean(arr))
print(np.max(arr))
print(np.min(arr))
print(np.std(arr))
```

```python
matrix = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print(np.sum(matrix))
print(np.sum(matrix, axis=0))  # column-wise
print(np.sum(matrix, axis=1))  # row-wise
```

#Numpy Day 2

Boolean masking

```python
import numpy as np

arr = np.array([5, 10, 15, 20, 25])

print(arr > 12)
print(arr[arr > 12])
print(arr[arr % 10 == 0])
```

```python
arr = np.array([3, 8, 12, 17, 21, 26])

filtered = arr[(arr > 10) & (arr < 25)]
print(filtered)
```

Broadcasting

```python
arr = np.array([1, 2, 3])
print(arr + 10)
print(arr * 5)
```

```python
matrix = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print(matrix + 10)
```

```python
matrix = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print(np.sum(matrix, axis=0))  # column-wise
print(np.sum(matrix, axis=1))  # row-wise

print(np.mean(matrix, axis=0))
print(np.mean(matrix, axis=1))
```

```python
arr1 = np.array([1, 2, 4])
arr2 = np.array([4, 5, 6])

print(np.concatenate((arr1, arr2)))
print(np.vstack((arr1, arr2)))
print(np.hstack((arr1, arr2)))
```
